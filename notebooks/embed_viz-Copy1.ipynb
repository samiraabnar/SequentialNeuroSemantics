```python
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import os

from tensorflow.contrib.tensorboard.plugins import projector

main_dir = '../log_root/'
model_dir = 'word2brain/word/intended/'
LOG_DIR =os.path.join(os.getcwd(),'output_embeddings',model_dir)
NAME_TO_VISUALISE_VARIABLE = "output_embeddings"


path_for_mnist_metadata =  os.path.join(LOG_DIR,'metadata.tsv')
```

```python
class Scan(object):
    def __init__(self, activations, timestamp, step, word=None, prev_word=None, next_word=None):
        self.activations = activations
        self.timestamp = timestamp
        self.word = word
        self.prev_word = prev_word
        self.next_word = next_word
        self.step = step


        
def read_and_prepare_data():
    predicted = np.load(main_dir+model_dir+"selectedBrain_10000_sigmoidlosspredicted_output.npy")
    target = np.load(main_dir+model_dir+"selectedBrain_10000_sigmoidlosstarget_output.npy")
    words = np.load(main_dir+model_dir+"selectedBrain_10000_sigmoidlosswords.npy")
   
    print(predicted[0].shape)
    print(target[0].shape)
    all_vectors = []
    if_target = []
    
    for i in np.arange(len(words)):
        all_vectors.append(predicted[0][i])
        if_target.append("pred")
        
    for i in np.arange(len(words)):
        all_vectors.append(target[i])
        if_target.append("target")
       
    all_vectors = np.asarray(np.asarray(all_vectors))
    

    return np.asarray(all_vectors), if_target, words
```

```python
all_vectors, if_target, words = read_and_prepare_data()
print(all_vectors.shape)
#lstm_embeddings, normalized_brain_scans, words = lstm_embeddings[:100], normalized_brain_scans[:100], words[:100]
embedding_var = tf.Variable(all_vectors, name=NAME_TO_VISUALISE_VARIABLE)
summary_writer = tf.summary.FileWriter(LOG_DIR)
```

```python
config = projector.ProjectorConfig()
embedding = config.embeddings.add()
embedding.tensor_name = embedding_var.name

# Specify where you find the metadata
embedding.metadata_path = path_for_mnist_metadata #'metadata.tsv'

# Specify where you find the sprite (we will create this later)
#embedding.sprite.image_path = path_for_mnist_sprites #'mnistdigits.png'
#embedding.sprite.single_image_dim.extend([28,28])

# Say that you want to visualise the embeddings
projector.visualize_embeddings(summary_writer, config)
```

```python
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())
saver = tf.train.Saver()
saver.save(sess, os.path.join(LOG_DIR, "model.ckpt"), 1)
```

```python
with open(path_for_mnist_metadata,'w') as f:
    f.write("Index\tLabel\tTarget\n")
    for index,label in enumerate(if_target):
        print(index,words[index % len(words)],label)
        f.write("%d\t%s\t%s\n" % (index,words[index % len(words)],label))
```
