```python
same_pos_accuracy = {}
for model in predicted.keys():
    print(model)
    same_pos_accuracy[model] = {}
    all_accuracies = []
    for key in pos_groups_index_dic.keys():
        print(key, len(pos_groups_index_dic[key]))
        selected_predictions = np.asarray(predicted[model])[pos_groups_index_dic[key]]
        selected_targets = np.asarray(target[model])[pos_groups_index_dic[key]]
        
        cosine_dists = distance.cdist(selected_predictions,selected_targets,'cosine')
        euc_dists =  distance.cdist(selected_predictions,selected_targets,'euclidean')
        c_acc,e_acc = eval(cosine_dists,euc_dists,model,len(pos_groups_index_dic[key]))
        print(c_acc,e_acc)
        same_pos_accuracy[model][key] = c_acc
```

```python
diff_pos_accuracy = {}
average_accuracy = {}
all_pos = list(pos_groups_index_dic.keys())
for model in predicted.keys():
    print(model)
    diff_pos_accuracy[model] = {}
    all_accuracies = []
    for i in np.arange(len(all_pos)):
        for j in np.arange(i,len(all_pos)):
                key_1 = all_pos[i]
                key_2 = all_pos[j]
                print(key_1,key_2, len(pos_groups_index_dic[key_1]),len(pos_groups_index_dic[key_2]))
                selected_predictions_1 = np.asarray(predicted[model])[pos_groups_index_dic[key_1]]
                selected_targets_1 = np.asarray(target[model])[pos_groups_index_dic[key_1]]
                selected_predictions_2 = np.asarray(predicted[model])[pos_groups_index_dic[key_2]]
                selected_targets_2 = np.asarray(target[model])[pos_groups_index_dic[key_2]]

                accuracies = []
                for pos1_sample_index in np.arange(len(selected_predictions_1)):
                    for pos2_sample_index in np.arange(len(selected_predictions_2)):
                        preds = [selected_predictions_1[pos1_sample_index],selected_predictions_2[pos2_sample_index]]
                        targets = [selected_targets_1[pos1_sample_index],selected_targets_2[pos2_sample_index]]
                        cosine_dists = distance.cdist(preds,targets,'cosine')
                        euc_dists =  distance.cdist(preds,targets,'euclidean')
                        c_acc,e_acc = eval(cosine_dists,euc_dists,model,2)

                        accuracies.append(c_acc)
                        all_accuracies.append(c_acc)
                print(np.mean(accuracies))
                diff_pos_accuracy[model][(key_1,key_2)] = np.mean(accuracies)
    average_accuracy[model] = np.mean(all_accuracies)
```

```python
keys = list(pos_groups_index_dic.keys())
keys_len = len(keys)
accuracy_mat_dic = {}
for model in predicted.keys():
    accuracy_mat_dic[model] = np.zeros(shape=(len(pos_groups_index_dic.keys()),len(pos_groups_index_dic.keys())))
    for accuracy_key in diff_pos_accuracy[model].keys():
        key_1 = accuracy_key[0]
        key_2 = accuracy_key[1]
        i = keys.index(key_1)
        j = keys.index(key_2)
        the_key = (key_1,key_2)
        if the_key not in  diff_pos_accuracy[model]:
            the_key = (key_2,key_1)
        accuracy_mat_dic[model][j][i] = diff_pos_accuracy[model][the_key]
    
```

```python


for model in accuracy_mat_dic.keys():
    accuracy_sum = 0
    accuracy_count = 0
    for key_1,key2 in diff_pos_accuracy[model].keys():
        if key_1 != key_2:
            i = keys.index(key_1)
            j = keys.index(key_2)
            accuracy_sum += ((len(pos_groups_index_dic[key_1])*len(pos_groups_index_dic[key_2]))/2) * accuracy_mat_dic[model][j,i]
            accuracy_count += ((len(pos_groups_index_dic[key_1])*len(pos_groups_index_dic[key_2]))/2)
    
    print(model,": Not same PoS Accuracy:", accuracy_sum/accuracy_count )  


  
for model in same_pos_accuracy.keys():
    accuracy_sum = 0
    accuracy_count = 0
    for key in same_pos_accuracy[model].keys():
        if (len(pos_groups_index_dic[key]) > 1):
            accuracy_sum += len(pos_groups_index_dic[key]) * same_pos_accuracy[model][key]
            accuracy_count += len(pos_groups_index_dic[key])

    print(model,": same PoS Accuracy:", accuracy_sum/accuracy_count )   
```

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

def plot_for_me(dataset,key_list,name,ax):
    mask = dataset == 0
    with sns.axes_style("white"):
        a = sns.heatmap(dataset,
                ax = ax,
        #           annot = True,
        #           fmt = 'f',
                mask=mask,
                xticklabels=key_list, 
                yticklabels=key_list, 
                linewidths=.5,
                cbar=False,
                cmap="YlGnBu",
                square=True,
                
                       )
        ax.set_xlabel(r'PoS Tags', fontname='Courier new',fontsize=10)
        ax.set_ylabel(r'PoS Tags', fontname='Courier new',fontsize=10)
        a.set_xticklabels(a.get_yticklabels(), rotation = 90,fontsize = 3)
        a.set_yticklabels(a.get_yticklabels(),fontsize = 3)

fig, (ax1,ax2,ax3,ax4) = plt.subplots(figsize=(5,20),ncols=4)
axes = [ax1,ax2,ax3,ax4]
for i in np.arange(len(predicted.keys())):
    model = list(predicted.keys())[i]
    plot_for_me(accuracy_mat_dic[model],keys,model,axes[i])

fig.savefig("contextual_acc"+".png", bbox_inches='tight', dpi=600)
```
